# Question 2

1. Add a new column "New_Age" in the DataFrame by adding 1 to values in existing "Age" column
2. Drop the column "SkinThickness" as it is not relevant to our analysis
3. Filter rows where patient's age is greater than or equal to 50 since such patients have high possibility of having diabetes. Use column New_Age. Incorporate another condition for filtering rows as “Outcome = 1”
4. Sort values in the DataFrame df4 by "Glucose" in descending order

```python
# Add a new column "New_Age" in the dataframe by adding 1 to values in existing "Age" column
df2['New_Age']=df['Age'] + 1
print(df2)
```

```python
df2.drop(['SkinThickness'], axis=1)
```

```python
df2.head()
```

```python
# Drop the column "SkinThickness" as it is not relevant to our analysis

df3 = df2.drop(['SkinThickness'], axis=1)
print(df3.head())
```

```python
df3['New_Age']>= 50
```

```python
# Filter rows where patient's age is greater than or equal to 50 since such patients have high possibility of having diabetes. 
# Use column New_Age

df3[df3['New_Age']>= 50]
```

```python
(df3['New_Age']>= 50) & (df3['Outcome'] == 1)
```

```python
# Incorporate another condition for filtering rows as Outcome is '1'

df4 = df3[(df3['New_Age']>= 50) & (df3['Outcome'] == 1)]
df4
```

```python
# Sort values in the dataFrame df4 by "Glucose" in descending order

df4.sort_values(by=['Glucose'], ascending = False)
```
